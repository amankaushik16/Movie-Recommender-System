```python
import numpy as np
import pandas as pd
```

```python
credits = pd.read_csv('tmdb_5000_credits.csv')
movies = pd.read_csv('tmdb_5000_movies.csv')
```

```python
movies.head()
```

```python
credits.head()
```

```python
movies = movies.merge(credits,on='title')
```

```python
movies.head()
```

```python
# COLOUMN RO BE CONSIDERED 
# 1. genres
# 2. id
# 3. keywords
# 4. title
# 5. overview
# 6. cast
# 7. crew

movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.head()
```

```python
# checking null values
movies.isnull().sum()
```

```python
# removing all the null values
movies.dropna(inplace=True)
```

```python
# check duplicate values 
movies.duplicated().sum()
```

```python
movies.iloc[0].genres
```

```python
# coverting String-of-List to List
import ast
ast.literal_eval('[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]')
```

```python
# '[{"id": 28, "name": "Action"}, {"id": 12, "name": "Adventure"}, {"id": 14, "name": "Fantasy"}, {"id": 878, "name": "Science Fiction"}]'
# converting upper line to below line 
# ['Action', 'Adventure', 'Fantasy', 'SciFi']

def convert(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter < 3:
            L.append(i['name'])
        counter+=1
    return L
     
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies.head()
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
movies.head()
```

```python
def convert3(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if counter != 3:
            L.append(i['name'])
            counter+=1
        else:
            break
        
    return L 
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
movies.head()
```

```python
def fetch_director(text):
    L = []
    counter = 0
    for i in ast.literal_eval(text):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(fetch_director)
```

```python
movies.head()
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies.head()
```

```python
# converting - "Aman Kaushik"
# to this - "AmanKaushik"

movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies.head()
```

```python
# merging all coloumns into one coloumn 
movies['tags'] =  movies['overview'] + movies['cast'] + movies['crew'] + movies['genres'] + movies['keywords']
```

```python
movies.head()
```

```python
# now removing all coloumns except 'TAGS'
# creating new dataframe including 'movie_id' , 'title' , 'tags'

new_df = movies[['movie_id','title','tags']]
```

```python
new_df
```

```python
# coverting 'TAGS' (list to string)
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df.head()
```

```python
import nltk
```

```python
from nltk.stem.porter import PorterStemmer
ps = PorterStemmer()
```

```python
# STEMING
# ['love', 'loved', 'loving'] -> ['love', 'love', 'love']

def stem(text):
    y=[]
    
    for i in text.split():
        y.append(ps.stem(i))
        
    return " ".join(y)
```

```python
new_df['tags'] = new_df['tags'].apply(stem)
```

```python
# list converted to string 
new_df['tags'][0]
```

```python
# convert every letter/word to lowercase
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
new_df['tags'][0]
```

```python
# VECTORIZATION 

from sklearn.feature_extraction.text import CountVectorizer 
cv = CountVectorizer(max_features=5000,stop_words='english')
```

```python
cv.fit_transform(new_df['tags']).toarray().shape
```

```python
vector = cv.fit_transform(new_df['tags']).toarray()
```

```python
vector[0]
```

```python
# most frequent 5000 words
cv.get_feature_names()
```

```python
# STEMING
# ['love', 'loved', 'loving'] -> ['love', 'love', 'love']

stem('In the 22nd century, a paraplegic Marine is dispatched to the moon Pandora on a unique mission, but becomes torn between following orders and protecting an alien civilization. SamWorthington ZoeSaldana SigourneyWeaver JamesCameron Action Adventure Fantasy cultureclash future spacewar')
```

```python
# movie-to-movie calculate "cosin" distance (angle between them)
# NOTE : distance inversely propotional to similarity

from sklearn.metrics.pairwise import cosine_similarity
```

```python
# passing all the vectors
cosine_similarity(vector)
```

```python
similarity = cosine_similarity(vector)
```

```python
# showing distance of one moving to all other 5000 movies
similarity[0]
```

```python
def recommend(movie):
    index = new_df[new_df['title'] == movie].index[0]
    distances = sorted(list(enumerate(similarity[index])),reverse=True,key = lambda x: x[1])
    for i in distances[1:6]:
        print(new_df.iloc[i[0]].title)
         
```

```python
recommend('Gandhi')
```
